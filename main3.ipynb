```python
#pegando dados de uma API , analisando tipo arquivo, fazendo um get apenas de alguns atributos e transfrmando em DATAFRAME
import requests
from pprint import pprint
```

```python
URL = 'https://restcountries.com/v3.1/all'
response = requests.get(URL)
response.status_code
```

```python
response.headers
```

```python
response.json()[:5]
```

```python
response_all = response.json()
response_all[0]
```

```python
response_all[0]['nome']
```

```python
# procurando pelo Brasil
for pais in response_all:
    if pais['name']['common'] == 'Brazil':
        brasil = pais
        print(brasil)
```

```python
pprint(brasil)
```

```python
URL_BRASIL = "https://restcountries.com/v3.1/name/brasil"

response_brasil = requests.get(URL_BRASIL).json()
response_brasil
```

```python
pprint(response_brasil)
```

```python
response_brasil[:5]
```

```python
lista_vazia= []

for keys in response_brasil:
    novo_dict = {
        'nome': keys ['name']['common'],
        'moeda': keys ['currencies']['BRL']['name'],
        'capital': keys ['capital'],
        'regiao': keys ['region']
        
    }
    lista_vazia.append(novo_dict)
novo_dict
```
